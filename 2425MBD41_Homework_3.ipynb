### 2. Get a dataframe (df3) with ORGANIZATION_TYPE containing "Transport"

```python
df1.head()
df3= df1[df1["ORGANIZATION_TYPE"].str.contains("Transport",case = False, na = False)]
df3["ORGANIZATION_TYPE"] = df3["ORGANIZATION_TYPE"].str[0:9]
df3
```

### 3. Create a new column OCCUPATION_TYPE_LEN in (df1) by counting the length of the string in OCCUPATION_TYPE

```python
df1["OCCUPATION_TYPE_LEN"]= df1["OCCUPATION_TYPE"].str.len()
df1
```

### 4. Show the number of records for each value of EDUCATION_TYPE in (df1)

```python
df1.head()#We check if education type exists in dF as Education Type
df1["EDUCATION_TYPE"] = df1["NAME_EDUCATION_TYPE"]
df1["EDUCATION_TYPE"].value_counts()
```

### 5. Get a new dataframe (df3) by sorting AMT_CREDIT in ascending order in (df1)

```python
df3 = df1.set_index("AMT_CREDIT")
df3
```

### 6. Get a dataframe (df3) with two columns (OCCUPATION_TYPE, ORGANIZATION_TYPE) and the duplicated records removed

```python
cols = ["OCCUPATION_TYPE","ORGANIZATION_TYPE"]
df3 = df1[cols]
df3_filtered=df3.drop_duplicates()
df3_filtered
```

### 7. Get a new dataframe (df3) with the average AMT_CREDIT by different OCCUPATION_TYPE in (df1)

```python
df3=df1.groupby('OCCUPATION_TYPE')["AMT_CREDIT"].mean()
df3=df3.apply(lambda x: f"{x:.2f}")
print(df3)
#The results were not very clear so i tried to format it with a function.
```

### 8. Get a new dataframe (df3) with the number (count) of previous applications for different SK_ID_CURR in (df2)

```python
df3=df2.groupby("SK_ID_CURR").size()
df3
#I was thinking of using .value_count but i gave me 1 each time so i moved to size()
```

### 9. Merge (df1) and (df2) and get a new dataframe (df3) with the percentage of previously approved applications for different OCCUPATION_TYPE

```python
df3=pd.merge(df1,df2, how="outer", on="SK_ID_CURR")
df_ok=df3[df3["NAME_CONTRACT_STATUS"]=="Approved"]

df3_ok=df3.groupby("OCCUPATION_TYPE")["NAME_CONTRACT_STATUS"].size()
df3=(df3_ok/df3_ok.sum())*100
percent_format = lambda x : f"{x:.1f}".replace(".", ",")+"%"
df3_formated = df3.apply(percent_format)
df3_formated
#I think it works, i divided df3.C by the sum of DFC and multiplied by 100. I did not found a way to apply a function That format to percentage
```
