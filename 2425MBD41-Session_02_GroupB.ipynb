```python
def sum_and_product_three_numbers(a, b, c):
  s = a+b+c
  p = a*b*c
  print('the sum is', s, 'and the product is', p)
  return s, p
```

```python
ss, pp = sum_and_product_three_numbers(2, 3, 4)
```

```python
sspp = sum_and_product_three_numbers(2, 3, 4)
```

```python
type(sspp)
```
